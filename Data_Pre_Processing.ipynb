```python
# Select columns
df_col_filtered = df.loc[:,['NP','ACR','BATH','BDSP','BLD','CONP','ELEP','FS','FULP','GASP','HFL',
                        'INSP','REFR','RMSP','RWAT','RWATPR','SINK','STOV','TEN','TOIL','VALP','VEH',
                        'WATP','YBL','FES','FPARC','HHL','HHT','HINCP','HUGCL','HUPAC','HUPAOC',
                         'HUPARC','KIT','LNGI','MULTG','MV','NOC','NPF','NPP','NR','NRC','PARTNER','PLM',
                        'PSF','R18','R60','R65','RESMODE','TAXP','WIF','WKEXREL','WORKSTAT']]
```

```python
# select rows based on TEN, the target data. The values have to be either
# 1.0: Owned with mortgage or loan (include home equity loans) or
# 2.0: Owed free and clear
df_col_row_filtered = df_col_filtered.loc[(df_col_filtered['TEN'] == 1.0) |
                                          (df_col_filtered['TEN'] == 2.0), :]
```

```python
df_col_row_filtered.info()
```

Based on the non-null information above, there are 4 non-null numbers:
1. 191909: All data points are non-null;
2. 187801: NaN should be mobile house / not a one-family house;
3. 140539: NaN should be non-family household;
4. 140070: NaN should include non-family household / same-sex married couple;

```python
# Firstly, for those variables affected by "non-family household"
# ['FPARC','NPF','WIF','WKEXREL','FES','WORKSTAT'], set 
# NaN as -1.

for var in ['FPARC','NPF','WIF','WKEXREL','FES','WORKSTAT']:
    df_col_row_filtered[var].fillna(-1,inplace=True)

# However, variables['FES','WORKSTAT'] are also affected
# by "same-sex married couple", set those values as 0.

df_col_row_filtered.loc[(df_col_row_filtered['FPARC']!=-1)&
                        (df_col_row_filtered['FES']==-1),['FES','WORKSTAT']] = 0

# Finally, the variable 'ACR' is affected by mobile house 
# / not a one-family house, set NaN as 0
df_col_row_filtered['ACR'].fillna(0,inplace=True)
```
